>Use crammer's rule to solve

$$
\begin{cases}
0.3x+0.52y+z = -0.01 \\
0.5x +y + 1.9z = 0.67\\
0.1x + 0.3y + 0.5z = -0.44

$$

Use Gauss elimination to solve

$$
\begin{cases}
3x-0.1y-0.2z = -7.85 \\
0.1x +7y - 0.3z = -19.3\\
0.3x - 0.2y + 10z = 71.4
\end{cases}
$$



```python
A = np.array([[3, -0.1, -0.2],[0.1, 7, -0.3],[0.3, -0.2, 10]])
b = np.array([[7.85],[-19.3],[71.4]])
# Eliminar Columna 1 (Operaciones Fila (2) y (3))
m_1 = A[1,0] / A[0, 0]
m_2 = A[2,0] / A[0, 0]
A[1, :] = A[1, :] - A[0, :] * m_1
b[1] = b[1] - b[0] * m_1
A[2, :] = A[2, :] - A[0, :] * m_2
b[2] = b[2] - b[0] * m_2
# Eliminar Columna 2
m_1 = A[2,1] / A[1, 1]
A[2, :] = A[2, :] - A[1, :] * m_1
b[2] = b[2] - b[1] * m_1
print(A)
x_3 = (b[2]) / A[2, 2]
x_2 = (b[1] - A[1, 2] * x_3) / A[1,1]
x_1 = (b[0] - A[0, 1] * x_2 - A[0, 2] * x_3) / A[0, 0]
print(x_3, x_2, x_1)
```

```python
def linear_system_solve(A, b):
    size = np.shape(A)
    if size[0] == size[1]:
        n = size[0]
        print(n)
        print(f'El tamaño de su matriz es {size}')
        vect_sol = np.zeros((n, 1))
        # Eliminación
        for i in range(n - 1):
            if A[i,i] == 0:
                raise ValueError("Pivote 0")
            for j in range(i + 1, n):
                factor = A[j, i] / A[i, i]
                A[j, :] = A[j, :] - A[i, :] * factor 
                b[j, 0] = b[j, 0] - b[i, 0] * factor 
        print(f"Sistema reducido: {A}, {b}")
        # solución x_n
        for i in range(n - 1, -1, -1):
            suma = 0
            for j in range(i + 1, n):
                suma += A[i, j] * vect_sol[j, 0]
            vect_sol[i, 0] = (b[i, 0] - suma) / A[i, i]
        print(f"El Vector Solución es: {vect_sol}")
        return vect_sol
    else:
        print(f'El tamaño de su matriz es {size}')
        return "Ingrese una matriz cuadrada nxn  (Ej: 3x3)"
```

```python
def gauss_naive(A, b):
    A = A.copy()
    b = b.copy()
    size = np.shape(A)
    if size[0] == size[1]:
        n = size[0]
        vect_sol = np.zeros((n, 1))
        for i in range(n - 1):
            for j in range(i + 1, n):
                alpha = A[j][i] / A[i][i] # <-- pivote
                A[j][:] = A[j][:] - A[i][:] * alpha
                b[j][0] = b[j][0] - b[i][0] * alpha
        # Susitución hacía atras
        for i in range(n - 1, -1, -1):
            suma = 0
            for j in range(i + 1, n):
                suma += A[i][j] * vect_sol[j][0]
            vect_sol[i][0] = (b[i][0] - suma) / A[i][i]
        print(f'el sistema reducido esta dado por: {A},\n {b}')
        print(f'El Vector Solución esta dado por: {vect_sol}')
    else:
        return "Este código solo permite matrices nxn (Ej: 2x2, 3x3, 4x4, etc)"    
```

```python
def gauss(A, b):
    
    A = A.copy()
    b = b.copy()
    size = np.shape(A)
    if size[0] == size[1]:
        n = size[0]
        vect_sol = np.zeros((n, 1)) 
                
        # Eliminación
        for i in range(n - 1):
            # Verificación Pivote
            if A[i][i] == 0:
                for k in range(i + 1, n):
                    if A[k][i] != 0:
                        #intercambio filas
                        A[[i, k]] = A[[k, i]]
                        b[[i, k]] = b[[k, i]]
                        break
                        
            for j in range(i + 1, n):
                alpha = A[j][i] / A[i][i] # <-- pivote
                A[j][:] = A[j][:] - A[i][:] * alpha
                b[j][0] = b[j][0] - b[i][0] * alpha
        # Susitución hacía atras
        for i in range(n - 1, -1, -1):
            suma = 0
            for j in range(i + 1, n):
                suma += A[i][j] * vect_sol[j][0]
            vect_sol[i][0] = (b[i][0] - suma) / A[i][i]
        print(f'el sistema reducido esta dado por: {A},\n {b}')
        print(f'El Vector Solución esta dado por: {vect_sol}')
    else:
        return "Este código solo permite matrices nxn (Ej: 2x2, 3x3, 4x4, etc)"    
```

```python
A = np.array([
    [0, 2, 1, 3],
    [1, -1, 2, 1],
    [2, 3, -1, 4],
    [1, 1, 1, 1]
], dtype=float)

b = np.array([
    [10],
    [3],
    [7],
    [4]
], dtype=float)
gauss(A, b)
```

>[!Example] Use the gauss-jordan technique to so]lve the same system

$$
\begin{cases}
3x-0.1y-0.2z = 7.85 \\
0.1x +7y - 0.3z = -19.3\\
0.3x - 0.2y + 10z = 71.4
\end{cases}
$$

```python
import numpy as np
```

```python
def gauss_jordan(A, b):
    A_sol = A.copy()
    b_sol = b.copy()
    size = np.shape(A)
    if size[0] == size[1]:
        n = size[0]
        
        if A_sol[0][0] != 0:
            # Eliminación inferior
            for i in range(n):
                b_sol[i][0] = b_sol[i][0] / A_sol[i][i] # tiene que ir primero, necesita utilizar el valor de A_sol sin modificar
                A_sol[i][:] = A_sol[i][:] / A_sol[i][i]

                for j in range(i + 1, n):
                    theta = A_sol[j][i] / A_sol[i][i]
                    A_sol[j][:] = A_sol[j][:] - A_sol[i][:] * theta
                    b_sol[j][0] = b_sol[j][0] - b_sol[i][0] * theta                

        print(A_sol, b_sol)
    else:
        return "Este código solo funciona para matrices cuadradas Ejemplo: 2x2, 3x3"            
```

```python
A = np.array([[3, -0.1, -0.2],[0.1, 7, -0.3],[0.3, -0.2, 10]], dtype = float)
b = np.array([[7.85],[-19.3],[71.4]], dtype = float)
gauss_jordan(A, b)
```

```python
np.linalg.solve(A, b)
```
