# Newton rapshon sistemas No-lineales
$$
u_{i+1} = u_i + (x_{i+1}-x_i)\dfrac{\partial u_i}{\partial x} + (y_{i+1}-y_i)\dfrac{\partial u_i}{\partial y}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 1.5
y0 = 3.5
# Funciones
u = lambda x, y:  x**2 + x * y  - 10
v = lambda x, y: y + 3 * x * y**2 - 57
# Funciones derivadas
u_x = lambda x, y: 2 * x + y
u_y = lambda x, y: x
v_x = lambda x, y: 3 * y**2 
v_y = lambda x, y: 1 + 6 * x * y
# Valores de x, y en el futuro
x_i = lambda x_0, y_0: x_0 - (u(x_0, y_0) * v_y(x_0, y_0) - v(x_0, y_0) * u_y(x_0, y_0)) / (u_x(x_0, y_0) * v_y(x_0, y_0) - u_y(x_0, y_0) * v_x(x_0, y_0))
y_i = lambda x_0, y_0: y_0 - (v(x_0, y_0) * u_x(x_0, y_0) - u(x_0, y_0) * v_x(x_0, y_0)) / (u_x(x_0, y_0) * v_y(x_0, y_0) - u_y(x_0, y_0) * v_x(x_0, y_0))

```

```python
x_i(x0, y0)
y_i(x0, y0)
for i in range(10):
    x0 = x_i(x0, y0)
    y0 = y_i(x0, y0)
    print(x0, y0)
```

```python
# Valores
x_0 = 4.5
x_1 = 5.5
x_2 = 5.0
f = lambda x: x**3 - 13 * x - 12
# Muller
h_0 = x_1 - x_0
h_1 = x_2 - x_1
delta_0 = (f(x_1) - f(x_0)) / (x_1 - x_0)
delta_1 = (f(x_2) - f(x_1)) / (x_2 - x_1)

print(h_0, h_1, delta_0, delta_1)
# Coeficientes
a = (delta_1 - delta_0) / (h_1 + h_0)
b = a * h_1 + delta_1
c = f(x_2)
print(a, b, c)
# quién es x4
x_3 = x_2 + - 2 * c / (b + np.sqrt(b**2 - 4 * a * c))
print(x_3)
```

```python
# Valores
x_0 = 4.5
x_1 = 5.5
x_2 = 5.0
for i in range(5):
    h_0 = x_1 - x_0
    h_1 = x_2 - x_1
    delta_0 = (f(x_1) - f(x_0)) / (x_1 - x_0)
    delta_1 = (f(x_2) - f(x_1)) / (x_2 - x_1)
    a = (delta_1 - delta_0) / (h_1 + h_0)
    b = a * h_1 + delta_1
    c = f(x_2)
    x_3 = x_2 + - 2 * c / (b + np.sqrt(b**2 - 4 * a * c))
    x_0 = x_1
    x_1 = x_2
    x_2 = x_3
    print(x_3)
```
